# Plotting DSC Data

This notebook will plot the DSC data for hydrogenated polystyrene, deuterated polystyrene, RRe-P3HT, RRa-P3HT, P3DDT and PQT-12 used in this work.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filenames = {
    'deut-ps':'../data/dsc/Deuterated_Polystyrene.csv',
    'hyd-ps':'../data/dsc/Hydrogenated_Polystyrene.csv',
    'rre':'../data/dsc/RReP3HT.csv',
    'rra':'../data/dsc/RRaP3HT.csv',
    'pqt':'../data/dsc/PQT12.csv',
    'p3ddt':'../data/dsc/P3DDT.csv',
    'rre2':'../data/dsc/RReP3HT2.csv',
}
```

```python
data = {}
for key in filenames:
    add_data = np.loadtxt(filenames[key], delimiter=',',encoding='utf-8-sig')
    add_data = add_data[np.where(add_data[:,-1]!=0)]
    data[key] = add_data
```

```python
colors = {
    'deut-ps':'black',
    'hyd-ps':'black',
    'rre':'firebrick',
    'rra':'darkorange',
    'pqt':'darkcyan',
    'p3ddt':'darkorchid',
    'rre2':'deeppink',
}

names = {
    'deut-ps':'Deuterated Polystyrene (PS-D8)',
    'hyd-ps':'Hydrogenated Polystyrene (PS-H8)',
    'rre':'RRe-P3HT',
    'rra':'RRa-P3HT',
    'pqt':'PQT-12',
    'p3ddt':'P3DDT',
    'rre2':'RRe-P3HT-2',
}
```

```python
plt.figure(figsize=(6,6))

for key, dsc in data.items():

    if 'rr' not in key:
        pass
        #plt.plot(dsc[:,1],dsc[:,2]*-1,linestyle='dotted',label=names[key],color=colors[key],linewidth=3)
    else:
        dsc1 = dsc[np.where((dsc[:,1]<260))]
        plt.plot(dsc1[:,1],dsc1[:,2]*-1,label=names[key],color=colors[key],linewidth=3)
    
plt.xlim(-40,260)
plt.ylim(-8,8)

plt.xlabel('Temperature (C)',fontsize=14)
plt.ylabel('Heat Flow Endothermic Up',fontsize=14)

plt.xticks(fontsize=14)
plt.yticks([],fontsize=14)

plt.annotate(s='Heating',xy=(100,7),xycoords='data',fontsize=14)
plt.annotate("",xy=(150,6.6), xycoords='data',xytext=(100,6.6), textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

plt.annotate(s='Cooling',xy=(100,-7.4),xycoords='data',fontsize=14)
plt.annotate("",xy=(100,-6.6), xycoords='data',xytext=(150,-6.6), textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))

plt.legend(fontsize=12,loc='upper left')

# adding in analysis information
t = 240.13 # RRe-P3HT melting temperature
plt.annotate(r"$T_{m,RRe-P3HT}=240.13 ^\circ$C",xy=(t,4.8), xycoords='data',xytext=(t-130,5.4), textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),fontsize=12)

# adding in analysis information
t = 8.26 # glass transition of RRa-P3HT
plt.annotate(r"$T_{g,RRa-P3HT}=8.26 ^\circ$C",xy=(t,1.3), xycoords='data',xytext=(t-40,3), textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),fontsize=12)

# adding in analysis information
t = 225.54 # RRe-P3HT melting temperature
plt.annotate(r"$T_{m,RRe-P3HT-2}=225.54 ^\circ$C",xy=(t,1.4), xycoords='data',xytext=(t-150,4), textcoords='data',arrowprops=dict(arrowstyle="->",connectionstyle="arc3"),fontsize=12)


plt.tight_layout()
plt.savefig('../data/dsc/p3ht_dsc.png',dpi=300)
```
